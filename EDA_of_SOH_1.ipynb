```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
import os
import glob
import seaborn as sns
```

loading dataset

```python
data = pd.read_csv('/content/ddd1.csv')
df = pd.DataFrame(data)
df
```

```python
print("Shape of data :", np.shape(data))
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 8))

plt.scatter(data['cycle'], data['SOH'])
plt.plot(data['cycle'], len(data['cycle'])*[0.75], color = 'r')

plt.ylabel('SOH', fontsize = 15)
plt.xlabel('cycles', fontsize = 15)
plt.title('SOH vs cycles', fontsize = 15)
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 8))

plt.scatter(data['cycle'], data['capacity'])
plt.plot(data['cycle'], len(data['cycle'])*[1.5], color = 'r')

plt.ylabel('capacity', fontsize = 15)
plt.xlabel('cycles', fontsize = 15)
plt.title('cycle vs capacity', fontsize = 15)
plt.show()
```
